### Problema 1.
Pokemon ha sido un juego de video y cartas que ha estado en el radar de muchos jóvenes y adultos, al grado que actualmente se ha visto un auge en la versión competitiva en los
juegos de video. Un pokemon cuenta con diferentes características que lo hacen único, entre ellas se encuentran el tipo de pokemon, el ataque, la defensa, la velocidad, la vida, entre
otros. En la versión competitiva se busca tener un equipo balanceado que pueda enfrentar a cualquier tipo de pokemon, por lo que se busca tener un equipo que tenga un balance

La problematica es encontrar un equipo de 6 pokemones que cumpla con las siguientes características:
1. Que el equipo tenga un balance entre ataque, defensa, velocidad y vida.
2. Que el equipo tenga un balance entre los tipos de pokemon para garantizar que pueda enfrentar a cualquier tipo de pokemon.

```python
# Librerías a utilizar
import pandas as pd
import kmapper as km
import matplotlib.pyplot as plt
plt.rcParams.update(plt.rcParamsDefault)
import numpy as np
import sklearn
```

### Exploración y preprocesamiento de datos

```python
# Leyendo la base de datos
df = pd.read_csv('Pokemon.csv')
df.head()
```

```python
# Copia del dataset original para su uso como diccionario
df_dic = pd.read_csv('Pokemon.csv')
```

```python
# Descripción general de las variables numéricas del set
df.describe()
```

```python
# Explorando la información general del set
df.info()
```

```python
# Reemplazando los datos nulos en la columna 'Type 2' por otra categoría referente a que el
# pokemon no tiene tipo secundario ('None')
for i in range(len(df)):
    if pd.isnull(df['Type 2'][i]):
        df.loc[i, 'Type 2'] = 'None'
```

```python
# Verificando que los valores nulos en 'Type 2' se hayan reemplazado
print(df.isnull().sum())
```

```python
# Eliminando la columna 'Name'
del(df['Name'])
```

```python
df_dic['Type 1'].unique()
```

```python
# Verificando las categorías de 'Type 1'
print(df['Type 1'].value_counts())
```

```python
# Verificando las categorías de 'Type 2'
print(df['Type 2'].value_counts())
```

```python
# Categorizando las columnas 'Type 1' y 'Type 2'
df['Type 1'] = df['Type 1'].astype('category').cat.codes
df['Type 2'] = df['Type 2'].astype('category').cat.codes
```

```python
# Categorizando la columna 'Legendary'
df['Legendary'].replace(to_replace=False, value=0, inplace=True)
df['Legendary'].replace(to_replace=True, value=1, inplace=True)
df
```

```python
# Calculando desviación estándar entre las variables HP, Attack, Defense y Speed esto con el objetivo de
# tener un índice de qué tan balanceado es un pokemon en sus habilidades
df['Balance'] = None
for i in range(len(df)):
  cols_for_sd = [df['Attack'][i], df['Defense'][i],df['HP'][i],df['Speed'][i]]
  df['Balance'][i] = int(np.std(cols_for_sd))
```

```python
# Transformación de datos de dataframe a numpy array para poder ser procesados en Mapper
df_map = df.to_numpy()
```

```python
# Inicializando Mapper
mapper = km.KeplerMapper(verbose=1)
```

```python
# Proyección de los datos sobre la variable 'Type 1' (tipo principal de un pokemon)
projected_data = mapper.fit_transform(df_map, projection=[2]) # X-Y axis
```

```python
# Generar las cubiertas | Segmentaación de los pokemones en 3 grupos principales
covering = km.Cover(n_cubes=3, perc_overlap=0.08)
```

```python
# Creación del grafo
graph = mapper.map(projected_data, df_map, clusterer=sklearn.cluster.KMeans(n_clusters=6), cover=covering)
```

```python
# Visualización del Mapper: Color con base en la variable 'Balance'
mapper.visualize(graph, path_html="Mapeo PT1.html",
                 title="Mapeo color Balance",
                 color_values = df_map[:,12],
                 color_function_name = "Balance",
                 node_color_function=np.array(['average','std','sum','max','min']))
```

```python
# Estadística descriptiva para el análisis por nodo
cubecluster=df.iloc[graph['nodes']['cube1_cluster0']]
cubecluster.describe()
```

```python
# Verificación de tipo principal predominante en cada nodo
cubecluster["Type 1"].value_counts()
```

```python
# Selección de filas del nodo representativo - Cluster 1
selected_1 = [df.iloc[indice] for indice in graph['nodes']['cube2_cluster4'] if df.loc[indice, 'Balance'] < 10]

# Creación de un nuevo DataFrame con las filas seleccionadas
df_sel1 = pd.DataFrame(selected_1)
```

```python
# Selección de filas del nodo representativo - Cluster 2
selected_2 = [df.iloc[indice] for indice in graph['nodes']['cube0_cluster5'] if df.loc[indice, 'Balance'] < 10]

# Creación de un nuevo DataFrame con las filas seleccionadas
df_sel2 = pd.DataFrame(selected_2)
```

```python
# Selección de filas del nodo representativo - Cluster 3
selected_3 = [df.iloc[indice] for indice in graph['nodes']['cube1_cluster2'] if df.loc[indice, 'Balance'] < 10]

# Creación de un nuevo DataFrame con las filas seleccionadas
df_sel3 = pd.DataFrame(selected_3)
```

```python
# Selección de filas del nodo representativo - Cluster 4
selected_4 = [df.iloc[indice] for indice in graph['nodes']['cube0_cluster4'] if df.loc[indice, 'Balance'] < 10]

# Creación de un nuevo DataFrame con las filas seleccionadas
df_sel4 = pd.DataFrame(selected_4)
```

```python
# Selección de filas del nodo representativo - Cluster 5
selected_5 = [df.iloc[indice] for indice in graph['nodes']['cube2_cluster0'] if df.loc[indice, 'Balance'] < 10]

# Creación de un nuevo DataFrame con las filas seleccionadas
df_sel5 = pd.DataFrame(selected_5)
```

```python
# Selección de filas del nodo representativo - Cluster 6
selected_6 = [df.iloc[indice] for indice in graph['nodes']['cube1_cluster0'] if df.loc[indice, 'Balance'] < 10]

# Creación de un nuevo DataFrame con las filas seleccionadas
df_sel6 = pd.DataFrame(selected_6)
```

```python
# Concatenando las filas seleccionadas
rows_selec = df_sel1, df_sel2, df_sel3, df_sel4, df_sel5, df_sel6
df_selected = pd.concat(rows_selec)
```

```python
# Recuperando las filas con el mayor puntaje 'en Total'
top_6_filas = df_selected.nlargest(6, 'Total')
dream_team = pd.DataFrame(top_6_filas)
```
